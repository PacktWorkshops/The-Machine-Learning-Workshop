```python
import unittest
import numpy as np
import pandas as pd
import numpy.testing as np_testing
import pandas.testing as pd_testing
import os
import import_ipynb
from exerciseClass import NN_Model

class Test(unittest.TestCase):

    def _dirname_if_file(self, filename):
        if os.path.isdir(filename):
            return filename
        else:
            return os.path.dirname(os.path.abspath(filename))

    def setUp(self):    
        import Exercise6_03_codeExtension
        self.exercise = Exercise6_03_codeExtension

        self.a = 1
        self.b = 0.56
        self.c = 1
        self.d = 1
        self.e = 1
        self.f = 0
        self.g  = 1
        self.h = -1
        self.i = 0.63

        self.model = NN_Model()
        
    def test_model(self):
        self.pred = self.pred = self.model.predict(season = self.a, age = self.b, childish = self.c, trauma = self.d, \
                    surgical = self.e, fevers = self.f, alcohol = self.g, smoking = self.h, sitting = self.i)

        np_testing.assert_almost_equal(self.pred, self.exercise.pred)

if __name__ == '__main__':
    unittest.main(argv=['first-arg-is-ignored'], exit=False)
```
